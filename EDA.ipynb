Provides functions to plot stacked bar charts and histograms with annotations for retention and churn data.

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:

        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
# Compute and plot the percentage distribution of churned vs. non-churned companies
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

```python
# Plot the distribution of electricity consumption over the past 12 months for all customers
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

### **SUB TASK 2: Hypothesis Verification: Evaluating the Correlation Between Price Sensitivity and Customer Churn**

```python
# Group by the number of products/services and count the number of customers
service_counts = client_df['nb_prod_act'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=service_counts.index, y=service_counts.values, palette='viridis')
plt.xlabel('Number of Services')
plt.ylabel('Number of Customers')
plt.title('Number of Customers with Different Numbers of Services')
plt.show()
```

```python
# Count the number of customers with and without gas service
gas_counts = client_df['has_gas'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x=gas_counts.index, y=gas_counts.values, palette='viridis')
plt.xlabel('Has Gas Service (1 = Yes, 0 = No)')
plt.ylabel('Number of Customers')
plt.title('Number of Customers with and without Gas Service')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
# Box Plot for Electricity and Gas Consumption
consumption_columns = ['cons_12m', 'cons_gas_12m', 'cons_last_month']
plt.figure(figsize=(10, 6))
sns.boxplot(data=client_df[consumption_columns])
plt.title('Box Plot of Electricity and Gas Consumption')
plt.show()
```

```python
# Histogram for Financial Metrics
financial_columns = ['margin_gross_pow_ele', 'margin_net_pow_ele', 'net_margin']
client_df[financial_columns].hist(bins=15, figsize=(15, 6), layout=(2, 2))
plt.suptitle('Histogram of Financial Metrics')
plt.show()
```

```python
# Identify numeric columns
numeric_cols = client_df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix for numeric columns
correlation_matrix = client_df[numeric_cols].corr()

# Assuming the last column is 'net_margin'
last_column_name = 'churn'
correlation_with_last_col = correlation_matrix[last_column_name].drop(last_column_name)

# Sort by absolute value of the correlation coefficients
sorted_correlation = correlation_with_last_col.abs().sort_values(ascending=False)

# Display the sorted correlation coefficients
print(sorted_correlation)
```

```python
# Plot the correlation coefficients
plt.figure(figsize=(12, 8))
sns.barplot(x=sorted_correlation.values, y=sorted_correlation.index, palette="viridis")
plt.title('Correlation Coefficients with Churn')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')
plt.show()
```
